```python
division=['a','b','c','d']
marks=[34,55,66,77]
avg_marks=[32,54,23,13]
width=0.3
index=np.arange(len(division))
#plt.xlim(40,100)
plt.bar(index+width,division,color='r',width=0.3,label='marks')
plt.bar(index-width,index,width=0.3,label='avg_marks')
plt.legend(loc='best')
plt.xticks(index,division)
plt.show()
```
